```python
import pandas as pd 
import numpy as np 

csvfile = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(csvfile)  
purchase_data["Age"] = pd.to_numeric(purchase_data["Age"])
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].unique()) 
print(f"Total amount of players: {total_players}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data['Item Name'].unique()) 
avg_price = purchase_data['Price'].mean()
number_purchases = purchase_data['Purchase ID'].count()
total_revanue = purchase_data['Price'].sum()
 
purchase_analysis = pd.DataFrame({"Number of unique items": [unique_items],
                                  "Average Price": [avg_price], 
                                  "Total Number of Purchases": [number_purchases], 
                                  "Total Revanue": [total_revanue]}) 
purchase_analysis["Average Price"] = purchase_analysis["Average Price"].map("${:.2f}".format)
purchase_analysis["Total Revanue"] = purchase_analysis["Total Revanue"].map("${:.2f}".format)
purchase_analysis.head() 
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_stats = purchase_data.groupby('Gender') 
total_gender = gender_stats["SN"].nunique()
total_players = len(purchase_data["SN"].value_counts())
percentage_players = total_gender / total_players * 100
gender_demographics = pd.DataFrame({"Percentage of Players": percentage_players, "Total Count": total_gender}) 
gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].map("{:.2f}%".format)
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purch_counts = gender_stats["Purchase ID"].count()
avgpurch_price = gender_stats["Price"].mean() 
totalpurch_value= gender_stats["Price"].sum() 

gender_purch_analysis = pd.DataFrame({"Purchase Count":purch_counts, 
                                      "Average Purchase Price":avgpurch_price, 
                                      "Total Purchase Value":totalpurch_value})
gender_purch_analysis["Average Purchase Price"] = gender_purch_analysis["Average Purchase Price"].map("{:.2f}%".format)
gender_purch_analysis["Total Purchase Value"] = gender_purch_analysis["Total Purchase Value"].map("${:.2f}".format)
gender_purch_analysis.head() 
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 99]
names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] 

purchase_data["Age Summary"] = pd.cut(purchase_data["Age"], bins, labels=names) 
age_grouped = purchase_data.groupby("Age Summary") 

total_grouped_count = age_grouped["SN"].nunique() 
grouped_percentage = total_grouped_count/total_players*100 

age_demographics = pd.DataFrame({"Percentage of Players":grouped_percentage}) 
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:.2f}%".format)
age_demographics.head(8)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
grouped_purch_count = age_grouped["SN"].count()
grouped_avg_price = age_grouped["Price"].mean()
grouped_purch_value = age_grouped["Price"].sum()
grouped_total_purch_avg = grouped_purch_value/total_grouped_count 

age_demographics = pd.DataFrame({"Purchase Count":grouped_purch_count, 
                                 "Average Price":grouped_avg_price, 
                                 "Total Purchase Value":grouped_purch_value, 
                                 "Average Purchase Total Per Person":grouped_total_purch_avg}) 
age_demographics["Average Price"] = age_demographics["Average Price"].map("${:.2f}".format)
age_demographics["Total Purchase Value"] = age_demographics["Total Purchase Value"].map("${:.2f}".format)
age_demographics["Average Purchase Total Per Person"] = age_demographics["Average Purchase Total Per Person"].map("${:.2f}".format)
age_demographics.head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data.groupby("SN") 
item_counts = top_spenders["Item Name"].count()
avg_purch_price = top_spenders["Price"].mean()
total_purch_value = top_spenders["Price"].sum()

top_spenders_df = pd.DataFrame({"Purchase Count":item_counts, 
                                "Average Purchase Price":avg_purch_price,
                                "Total Purchase Value":total_purch_value}) 

top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:.2f}".format)

top_spenders_df.sort_values("Total Purchase Value",ascending=False).head(5)


```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items = purchase_data.groupby("Item Name")

item_id = popular_items["Item ID"].unique()
item_purch_count = popular_items["Item Name"].count()
item_price = popular_items["Price"].unique()
item_purch_value = popular_items["Price"].sum() 

popular_items_df = pd.DataFrame({"Item ID":item_id, 
                                 "Item Purchase Count":item_purch_count, 
                                 "Price":item_price, 
                                 "Total Purchase Value":item_purch_value})

popular_items_df["Total Purchase Value Formatted"] = popular_items_df["Total Purchase Value"].map("${:.2f}".format)

popular_items_df.sort_values("Item Purchase Count",ascending=False).head()
```
